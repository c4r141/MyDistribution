```python
import scipy
import numpy as np
from scipy.stats import pareto
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = 3.
m = 2.
```

```python
Sample = (np.random.pareto(a,1000)+1) *m
```

```python
Sample[0:5]
```

```python
ax= sns.distplot(Sample,
            kde=True,
            bins=10,
            color='skyblue',
            hist_kws= {'linewidth':15, 'alpha':1, 'color': 'green'})

ax.set(xlabel ="Pareto Distribution", ylabel= "Frequency")
```
